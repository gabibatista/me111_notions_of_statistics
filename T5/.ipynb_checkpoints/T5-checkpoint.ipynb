2.Calcule a média, a mediana, o mínimo, o máximo do tempo para completar os testes usando os dados das suas 3 tentativas.

```r
meusTempos = c(meuTeste$Tempo1, meuTeste$Tempo2, meuTeste$Tempo3)
minhasIdades = c(meuTeste$Idade1, meuTeste$Idade2, meuTeste$Idade3)

media = mean(meusTempos)
mediana = median(meusTempos)
minimo = min(meusTempos)
maximo = max(meusTempos)
media_idades = mean(minhasIdades)

print(sprintf('Dentre as minhas 3 tentativas, o tempo médio para completar o teste é %s segundos; a mediana é %s.',
        media, mediana))
print(sprintf('O menor e o maior tempo foram %s e %s segundos, respectivamente. Minha idade mental média é %s anos.',
         minimo, maximo, media_idades))
```

4.Cada aluno realizou 3 tentativas. Para definir qual o aluno mais rápido, primeiro podemos resumir as medidas das 3 tentativas de cada aluno usando: a mediana, a média, o mínimo ou o máximo. Preencha as frases a seguir. Qual medida para resumir os dados você acha que é a melhor nesse caso e por que?

```r
## Por exemplo, vamos calcular o máximo dos tempos
## para todos os alunos usando a função 'apply'
tempos = sinestesia[, c("Tempo1", "Tempo2", "Tempo3")]
max_tempos = apply(tempos, MARGIN = 1, FUN = max)
min_tempos = apply(tempos, MARGIN = 1, FUN = min)
avg_tempos = apply(tempos, MARGIN = 1, FUN = mean)
med_tempos = apply(tempos, MARGIN = 1, FUN = median)

print(sprintf('Se usarmos o tempo máximo de cada um, o menor tempo foi %s do aluno RA %s.', 
              min(max_tempos), sinestesia$RA[match(min(max_tempos), max_tempos)]))
print(sprintf('Se usarmos o mínimo, o menor tempo foi %s do aluno RA %s.', 
              min(min_tempos), sinestesia$RA[match(min(min_tempos), min_tempos)]))
print(sprintf('Se usarmos a média, o menor tempo foi %s do aluno RA %s.', 
              min(avg_tempos), sinestesia$RA[match(min(avg_tempos), avg_tempos)]))
print(sprintf('Se usarmos a mediana, o menor tempo foi %s do aluno RA %s.', 
              min(med_tempos), sinestesia$RA[match(min(med_tempos), med_tempos)]))
```

<b>Resposta:</b> Nesse caso, os métodos utilizados não alteram o resultado final, então acredito que tanto faz o método escolhido. Eu escolheria buscar o mínimo entre a média dos valores porque leva em consideração todos os valores.

5.Escolha uma estatística para resumir os dados de cada aluno (média, mediana, mínimo ou máximo). Usando a estatística escolhida, qual o maior tempo entre todos os alunos? E qual a menor idade mental?

```r
tempos = sinestesia[, c("Tempo1", "Tempo2", "Tempo3")]
idades = sinestesia[, c("Idade1", "Idade2", "Idade3")]

avg_tempos = apply(tempos, MARGIN = 1, FUN = mean)
avg_idades = apply(idades, MARGIN = 1, FUN = mean)

print(sprintf('O maior tempo entre todos os alunos é %s do aluno de RA %s.', 
              max(avg_tempos), sinestesia$RA[match(max(avg_tempos), avg_tempos)]))

print(sprintf('A menor idade mental entre todos os alunos é %s do aluno de RA %s.', 
              min(avg_idades), sinestesia$RA[match(min(avg_idades), avg_idades)]))
```

Nas questões a seguir, considere apenas a mediana dos tempos e das idades das 3 tentativas de cada aluno.


6.Construa dois gráficos: uma para descrever a distribuição das medianas dos tempos e outro para a distribuição das medianas das idades.

```r
med_tempos = apply(tempos, MARGIN = 1, FUN = median)
med_idades = apply(idades, MARGIN = 1, FUN = median)

ggplot() + geom_density(aes(x=med_tempos, group=1)) + ggtitle('Distribuição das medianas dos tempos')
ggplot() + geom_density(aes(x=med_idades, group=1)) + ggtitle('Distribuição das medianas das idades')
```

7.Indique a mediana dos tempos e das idades das suas 3 tentativas nos respectivos gráficos usando uma cor diferente.

```r
ggplot() + geom_density(aes(x=med_tempos, group=1)) + ggtitle('Distribuição das medianas dos tempos') +
          annotate("point", x = mediana, y=0.018, colour = "blue")
ggplot() + geom_density(aes(x=med_idades, group=1)) + ggtitle('Distribuição das medianas das idades') +
         annotate("point", x = median(minhasIdades), y=0.027, colour = "blue")
```

8.Quantos alunos têm tempo de teste maior do que o seu? E quantos têm tempo de teste menor do que o seu?

```r
tempo_maior = length(subset(med_tempos, med_tempos > median(meusTempos)))
tempo_menor = length(subset(med_tempos, med_tempos < median(meusTempos)))

print(sprintf('%s alunos têm tempo de teste maior que o meu enquanto %s alunos têm tempo menor que o meu', 
              tempo_maior, tempo_menor))
```

9.Quantos alunos têm idade mental maior do que a sua? E quantos têm idade mental menor que a sua?

```r
idade_maior = length(subset(med_idades, med_idades > median(minhasIdades)))
idade_menor = length(subset(med_idades, med_idades < median(minhasIdades)))

print(sprintf('%s alunos têm idade mental maior que a minha enquanto %s alunos têm idade mental menor que a minha', 
              idade_maior, tempo_menor))
```
