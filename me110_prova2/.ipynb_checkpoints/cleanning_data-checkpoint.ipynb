```r
data = read.csv("RA197689obitos-estados_raw.csv")
```

```r
head(data)
```

```r
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="< 9") "<= 39" else "")
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="10 - 19") "<= 39" else x[7])
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="20 - 29") "<= 39" else x[7])
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="30 - 39") "<= 39" else x[7])

data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="40 - 49") "40 - 79" else x[7])
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="50 - 59") "40 - 79" else x[7])
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="60 - 69") "40 - 79" else x[7])
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="70 - 79") "40 - 79" else x[7])
    
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[7]=="") ">= 80" else x[7])
data$faixa_etaria_agg <- apply(data, 1, FUN = function(x) if(x[4]=="") "" else x[7])
```

```r
data$tipo_doenca_agg <- apply(data, 1, FUN = function(x)
    if(x[2]=="COVID") "COVID-19" else "")

data$tipo_doenca_agg <- apply(data, 1, FUN = function(x) 
    if(x[2]=="SRAG") "Outras doenças respiratórias" else x[8])
data$tipo_doenca_agg <- apply(data, 1, FUN = function(x) 
    if(x[2]=="PNEUMONIA") "Outras doenças respiratórias" else x[8])
data$tipo_doenca_agg <- apply(data, 1, FUN = function(x) 
    if(x[2]=="INSUFICIENCIA_RESPIRATORIA") "Outras doenças respiratórias" else x[8])
data$tipo_doenca_agg <- apply(data, 1, FUN = function(x) 
    if(x[2]=="SEPTICEMIA") "Outras doenças respiratórias" else x[8])
        
data$tipo_doenca_agg <- apply(data, 1, FUN = function(x)
    if(x[2]=="INDETERMINADA") "Outras doenças" else x[8])
data$tipo_doenca_agg <- apply(data, 1, FUN = function(x)
    if(x[2]=="OUTRAS") "Outras doenças" else x[8])
```

```r
data$faixa_etaria <- data$faixa_etaria_agg
data$tipo_doenca <- data$tipo_doenca_agg
data <- data[ , -which(names(data) %in% c("faixa_etaria_agg","tipo_doenca_agg"))]
```

```r
data = data[data$faixa_etaria != "", ]
```
