# Questão 2

```r
for (est in as.list(unique(df$estado))){
    print(sprintf('------------ %s', est))
    print('--- casosNovos')
    print(summary(subset(df, estado==est)$casosNovos))
    print('--- obitosNovos')
    print(summary(subset(df, estado==est)$obitosNovos))
}
```

```r
for (reg in as.list(unique(df$regiao))){
    print(sprintf('------------ %s', reg))
    print('--- casosNovos')
    print(summary(subset(df, regiao==reg)$casosNovos))
    print('--- obitosNovos')
    print(summary(subset(df, regiao==reg)$obitosNovos))
}
```
